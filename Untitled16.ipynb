# Spiltting the data into Training and Test

```python
df.info()
X = df.drop("mpg",axis=1) #independent variables or features
Y = df['mpg'] #dependent variable or target feature
```

# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
X_train_std
```

```python
pickle.dump(sc,open('E:\Projects\Python\Fuel Efficiency\scaler.pkl','wb'))
```

# Build our Model and Train

```python
# Intializing the ANN
model = Sequential()

# Adding the input layer
model.add(Dense(units=64,activation='relu',input_shape=[7]))

#Adding second layer
model.add(Dense(units=64,activation='relu'))

#Adding third layer
model.add(Dense(units=64,activation='relu'))

#Adding output layer
model.add(Dense(units=1))
```

```python
# Compiling the ANN
model.compile(loss='mean_squared_error',optimizer='Adam',metrics=['mean_squared_error','mean_absolute_error'])
```

```python
model.summary()
```

# Training the model

```python
model.fit(X_train_std,Y_train,batch_size=100,epochs=1000)
```

```python
model.evaluate(X_train_std,Y_train)
```

```python
model.evaluate(X_test_std,Y_test)
```

```python
prediction = model.predict(X_test_std)
```
